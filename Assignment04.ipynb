# 2. LeetCode 232. Implement Queue using Stacks

```python
class MyQueue(object):

    def __init__(self):
        self.stack_in = []   # 들어오는 요소용
        self.stack_out = []  # 나갈 때 사용하는 용

    def push(self, x):
        self.stack_in.append(x)

    def pop(self):
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        return self.stack_out.pop()

    def peek(self):
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        return self.stack_out[-1]

    def empty(self):
        return not self.stack_in and not self.stack_out
```

__init__
* stack_in: 새 요소를 받을 때 스택
* stack_out: 새 요소를 꺼낼 때 사용하는 스택

push
* 요소 x를 큐의 뒤에 추가

pop
* 큐의 앞 요소를 제거하고 반환.
* stack_out이 비어 있으면 stack_in의 요소를 모두 옮겨서 순서 반전

peak
* 큐의 앞 요소를 반환
* stack_out을 채우고, 마지막 요소 확인

empty
* 큐가 비어있는지 확인
* 두 스택이 모두 비어있으면 True 반환

```python
class ListQueue:
    def __init__(self):
        self.__queue = []

    def enqueue(self, x):
        self.__queue.insert(0, x)     # 앞에 삽입

    def dequeue(self):
        return self.__queue.pop()     # 끝에서 제거

    def front(self):
        return self.__queue[-1]       # 끝 요소 반환

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()
```

* 앞부분은 큐에 차례대로 넣고, 뒷부분은 앞에서 하나씩 꺼내며 비교

```python
from collections import deque

def isMirrorString(s):
    q = deque()
    i = 0
    while i < len(s) and s[i] != '$':
        q.append(s[i])
        i += 1
    i += 1

    while i < len(s):
        if not q or q.pop() != s[i]:
            return False
        i += 1

    return not q
```

### 4.
* push(x): 큐에 넣음.
* pop(): 하나만 남을 때까지 다른 큐로 옮기고, 마지막 하나만 꺼냄.

```python
class StackUsingQueues:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        while len(self.q1) > 1:
            self.q2.append(self.q1.pop(0))
        top = self.q1.pop(0)
        self.q1, self.q2 = self.q2, self.q1
        return top
```

### 5.
* enqueue(x): s1에 넣기
* dequeue(): s2가 비어있으면 s1을 모두 pop해서 s2에 push → s2는 거꾸로 된 순서니까 맨 앞처럼 꺼낼 수 있음

```python
class QueueUsingStacks:
    def __init__(self):
        self.s1 = []
        self.s2 = []

    def enqueue(self, x):
        self.s1.append(x)

    def dequeue(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop() if self.s2 else None
```

### 8.
* addFront(x)	insert(0, x) → 앞에 삽입
* addRear(x)	append(x) → 뒤에 삽입
* deleteFront()	pop(0) → 앞에서 삭제
* deleteRear()	pop() → 뒤에서 삭제
* getFront()	list[0]
* getRear()	list[-1]

```python
class Deque:
    def __init__(self):
        self.__deque = []

    def addFront(self, x):
        self.__deque.insert(0, x)  # 앞에 삽입

    def addRear(self, x):
        self.__deque.append(x)     # 뒤에 삽입

    def deleteFront(self):
        if self.isEmpty():
            return None
        return self.__deque.pop(0)  # 앞에서 삭제

    def deleteRear(self):
        if self.isEmpty():
            return None
        return self.__deque.pop()   # 뒤에서 삭제

    def getFront(self):
        if self.isEmpty():
            return None
        return self.__deque[0]      # 앞 요소 확인

    def getRear(self):
        if self.isEmpty():
            return None
        return self.__deque[-1]     # 뒤 요소 확인

    def isEmpty(self) -> bool:
        return len(self.__deque) == 0

    def clear(self):
        self.__deque.clear()

    def printDeque(self):
        print("Deque from front:", end=' ')
        for item in self.__deque:
            print(item, end=' ')
        print()
```
